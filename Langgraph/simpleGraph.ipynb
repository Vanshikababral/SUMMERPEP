```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph

from IPython.display import display, Image
```

```python
class AgentState(TypedDict):
    msg : str
```

```python
def greeting_node(state: AgentState) -> AgentState:
    state['msg'] = "Hello, World" #passing value to the dict 
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("Hello World", greeting_node)
graph.set_entry_point("Hello World")
graph.set_finish_point("Hello World")
```

```python
app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
app.invoke({
    "msg" : "Hello"
})
```

### Graph 2

```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name : str
    age : int
    marks_list : List[int]
    msg : str
```

```python
def calculation(state: AgentState) -> AgentState:
    state['msg'] = f"Hey {state['name']} , your age is {state['age']}, sum is {sum(state['marks_list'])}."
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("records", calculation)
graph.set_entry_point("records")
graph.set_finish_point("records")
```

```python
app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
app.invoke({
        "name" : "Vanshika",
        "age" : 24,
        "marks_list" : [23, 35, 34]
    })

```
